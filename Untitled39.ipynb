```python
# Given Data
x1, x2, x3 = 2, 1, 3
w1, w2, w3 = 0.5, -1, 2
b = -2
lr = 0.1
target = 1


# Step 1 : Net Input
z = (w1*x1) + (w2*x2) + (w3*x3) + b
print("Net input z =", z)


# Step 2 : Activation
if z >= 0:
    y_pred = 1
else:
    y_pred = 0

print("Predicted output =", y_pred)


# Step 3 : Error
error = target - y_pred
print("Error =", error)


# Step 4 : Weight Update
w1 = w1 + lr * error * x1
w2 = w2 + lr * error * x2
w3 = w3 + lr * error * x3

print("Updated weights:", w1, w2, w3)


#  Step 5 : Bias Update
b = b + lr * error
print("Updated bias =", b)


# Step 6 : New Prediction
z_new = (w1*x1) + (w2*x2) + (w3*x3) + b
print("New z =", z_new)

if z_new >= 0:
    new_pred = 1
else:
    new_pred = 0

print("New Prediction =", new_pred)
```
